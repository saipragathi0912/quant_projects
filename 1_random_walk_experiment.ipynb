```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
num_of_iterations = 100
initial_value = 100
values = np.ones(num_of_iterations)
values[0] = initial_value
for i in range(1,num_of_iterations):
    if random.uniform(0,1) > 0.5:
        values[i] = values[i-1]*1.01
    else:
        values[i] = values[i-1]*0.99
```

```python
plt.plot(values)
```
